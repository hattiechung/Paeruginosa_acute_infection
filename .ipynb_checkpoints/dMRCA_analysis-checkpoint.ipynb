# helper functions

```python
import toytree, toyplot, re, copy 
import toyplot.pdf
import numpy as np
import matplotlib.pyplot as plt

ALL_COLORS = ["rgba(33,146,164,1)", #2192a4
                "rgba(0,3,128,1)",  #000380
                "rgba(239,133,91,1)", #ef855b
                "rgba(50%,50%,50%,1)"]

ALL_COLORS_HASH = ['#2192a4','#000380','#ef855b']

ALL_COLORS_SCALED = [(0.1294,0.5725,0.6431),
                     (0,0.0118,0.5020),
                     (0.9373,0.5216,0.3569)]

def natural_sort(l): 
    convert = lambda text: int(text) if text.isdigit() else text.lower() 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)

def get_tree(tfile): 
    with open(tfile,'r') as tfile: 
        tree = tfile.read().replace('\n','')
    return tree

def get_unique_sample_strings(treestr): 
    namebit = '\w+(?=\-)'
    matches = re.findall(namebit, treestr)
    samples = list(set(matches))
    samples = natural_sort(samples)
    colordict = dict(zip(samples,ALL_COLORS[0:len(samples)]))
    return samples, colordict

def get_tip_colors(colordict,treeobj):
    colorlist = [colordict[t[0:t.find('-')]] if t.find('-') != -1 else \
                 "black" for t in treeobj.get_tip_labels()]
    return colorlist

def get_node_colors(colordict,treeobj): 
    colorlist = [colordict[t[0:t.find('-')]] if t.find('-') != -1 else \
                 "white" for t in treeobj.get_node_dict().values()]
    return colorlist

def get_leaves_rename_dict(): 
    samplenamedict = {'05SP1A':'A1',
                         '45SP1':'E1', 
                         '13SP1':'B1',
                         '13SP3':'B4',
                         '42SP1':'F0',
                         '42SP2':'F5',
                         '42ST1':'F6',
                         '14SP1':'C0',
                         '14SP3':'C10',
                         '46SP1B':'G0',
                         '46SP3':'G10',
                         '09SP1':'D0',
                         '09SP4':'D11',
                         '09ST1':'D5',
                         '10SP1b':'H0',
                         '10SP4b':'H11',
                         '01SP9':'I39',
                         '01SP10':'I50'} 
    return samplenamedict

def rename_tree_leaves(tobj): 
    t_renamed = copy.deepcopy(tobj)
    samplenamedict = get_leaves_rename_dict() 
    namebit = '\w+(?=\-)'
    for leaf in t_renamed.treenode.iter_leaves():
        if '-' in leaf.name: 
            leaf_sample = re.findall(namebit, leaf.name)
            newname = samplenamedict[leaf_sample[0]]
#             newname_isol = re.sub(leaf_sample[0],newname,leaf.name)
            newname_isol0 = re.sub(leaf_sample[0],'',leaf.name) # remove first part of isolate name
            newname_isol = re.sub('-','',newname_isol0) # remove first part of isolate name
            leaf.name = newname_isol
    return t_renamed

def get_plot_attr(patient):
    patient_plot_width = {'P13':150, #90, 
                          'P14':125, #170,
                          'P42':170,
                          'P10':400,
                            'P46':200,
                          'P45':150,
                         'P09':150,
                         'P01':180,
                         'P05':140}
    patient_plot_height = {'P45':200, #90, 
                           'P13':500,
                          'P14':500, #170,
                          'P42':500,
                          'P10':500,
                         'P46':500,
                          'P45':300,
                         'P09':500,
                         'P01':400,
                         'P05':250}
    patient_tip_font = {'P13':"8px",
                        'P14':"9px",
                        'P45':"8px", #"8px"
                       'P42':"7px",
                       'P10':"8px",
                       'P46':"7px",
                       "P09":"6px",
                       'P01':"6px",
                       'P05':"6px"} #"8px"
    try: 
        wid = patient_plot_width[patient]
        hgt = patient_plot_height[patient]
        fsize = patient_tip_font[patient]
    except KeyError: 
        wid = 400
        hgt = 500
        fsize = "9px"
    return wid, hgt, fsize

def move_outgroup_to_bottom(treeh): 
    outgroup_idx = treeh.get_tip_labels().index('Outgroup')
    all_coords = treeh.get_tip_coordinates()
    new_coords = all_coords.copy()

    # move index of all leaves up by one along y-coord
    for i in range(0,outgroup_idx):
        cur_coord = all_coords[i]
        new_coord = [cur_coord[0], cur_coord[1]+1]
        new_coords[i] = new_coord

    new_coords[outgroup_idx,1] = 0

    for idx, coords in zip(range(len(new_coords)),new_coords): 
        treeh.get_tip_coordinates()[idx] = coords
    return treeh, new_coords

def get_root_to_tip_dists(tobj,samplestr,leaf_to_sample):
    distances = {key: [] for key in samplestr}
    for leaf in tobj.get_tip_labels():
        dist = tobj.treenode.get_distance(leaf)
        if leaf in leaf_to_sample: 
            distances[leaf_to_sample[leaf]].append(dist) 
    mean_distances = []
    std_distances = []
    for sample in samplestr: 
        dist_mean = np.mean(distances[sample])
        mean_distances.append(dist_mean)
        std_distances.append(np.std(distances[sample]))
    return mean_distances, distances, std_distances

def get_leafname_to_sample(tobj,permute=False): 
#     namebit = '\w+(?=\-)'
    namebit = '[0-9]+SP\w+(?=\-)'
    leaves = tobj.get_tip_labels()
    leaf_sample = {}
    for leaf in leaves:
        matches = re.findall(namebit, leaf)
        if len(matches) == 1: 
            lsample = matches[0]
            leaf_sample[leaf] = lsample
    if permute: 
        orig_keys = list(leaf_sample.keys())
        keys = random.permutation(orig_keys)
        leaf_sample = dict(zip(keys, leaf_sample.values()))
    return leaf_sample

def get_pvalues(tobj,samplestr):
    # version 2: calculate the difference of <dMRCA> between time point 1 and time point 2
    leafsamples_obs = get_leafname_to_sample(tobj, permute=False)
    obs_vals, dists_obs, obs_vals_std = get_root_to_tip_dists(tobj,samplestr,leafsamples_obs)
    obs_diff = obs_vals[1]-obs_vals[0]

    i = 0
    null_vals = [] #np.empty([0,1],dtype=float)
    while i<numtrials: 
        leafsamples_null_i = get_leafname_to_sample(tobj, permute=True)
        null_i, dists_null_i, null_i_std = get_root_to_tip_dists(tobj, samplestr, leafsamples_null_i)
        null_diff = null_i[1]-null_i[0]
#         null_vals = np.vstack([null_vals,null_diff])
        null_vals.append(null_diff)
        i+=1

    pval_s = float(sum(obs_diff>null_vals))/numtrials
    print(obs_diff,pval_s,1-pval_s)
    
    return obs_vals, obs_vals_std, obs_diff, null_vals

def get_leaf_to_sample(tobj):
    namebit = '\w+(?=\-)'
    leaves = tobj.get_tip_labels()
    leaf_sample = {}
    for leaf in leaves: 
        matches = re.findall(namebit, leaf)
        if len(matches) == 1: 
            lsample = matches[0]
            leaf_sample[leaf] = lsample
    return leaf_sample

def get_root_to_tip_dists_all(tobj,samplestr):
    from scipy import stats
    distances = {key: [] for key in samplestr}
    leaf_to_sample = get_leaf_to_sample(tobj)
    for leaf in tobj.get_tip_labels():
        dist = tobj.treenode.get_distance(leaf)
        if leaf in leaf_to_sample:
            try: 
                distances[leaf_to_sample[leaf]].append(dist)
            except KeyError:
                continue
    mean_distances = []
    std_distances = []
    for sample in samplestr: 
        print(distances[sample])
        dist_mean = np.mean(distances[sample])
        mean_distances.append(dist_mean)
#         std_distances.append(np.std(distances[sample]))
        std_distances.append(stats.sem(distances[sample]))
    return mean_distances, distances, std_distances

def get_mean_MRCA(tobj,samplestr):
    leafsamples_obs = get_leafname_to_sample(tobj, permute=False)
    obs_vals, dists_obs, obs_vals_std = get_root_to_tip_dists_all(tobj,samplestr)

    return obs_vals, obs_vals_std
```

```python
def load_patient_tree(patient):
    tfile = '%s/%s_singleSNP.tree' %(patient, patient)
    newick = get_tree(tfile)
    treeh = toytree.tree(newick)
    treeh.ladderize()
    treeh.root("Outgroup") 
    treeh.rotate_node(names='Outgroup')
    
    return treeh, newick
```

```python
# test for significant differences between root-to-tip distances of sample 1 vs. sample 2
import itertools
from numpy import random
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pylab as P

numtrials = 1000

def get_sample_hbar_rename_dict(): 
    samplenamedict = {'05SP1A':'Day 1',
                         '45SP1':'Day 1', 
                         '13SP1':'Day 1',
                         '13SP3':'Day 5',
                         '42SP1':'Day 1',
                         '42SP2':'Day 6',
                         '14SP1':'Day 1',
                         '14SP3':'Day 11',
                         '46SP1B':'Day 1',
                         '46SP3':'Day 11',
                         '09SP1':'Day 1',
                         '09SP4':'Day 12',
                         '10SP1b':'Day 1',
                         '10SP4b':'Day 12',
                         '01SP9':'Day 1',
                         '01SP10':'Day 12'} 
    return samplenamedict

def get_barplot_size(pat,plottype): 
    pat_vplot = {'P10':(1.5,2),
               'P13':(1.5,2)}
    pat_hplot = {'P10':(1.5,2),
                'P09':(1.5,2), 
                'P13':(1.5,2), 
                'P14':(1.5,2), 
                'P46':(1.5,2),
                 'P01':(1.5,2),
                'P42':(1.5,2)}
    if plottype == 'vertical': 
        try: 
            figsize = pat_vplot[pat]
        except KeyError: 
            figsize = (3,6)
    elif plottype == 'horizontal': 
        try: 
            figsize = pat_hplot[pat]
        except KeyError: 
            figsize = (6,4)
    return figsize
    
def plot_hbar(obs,stds,samplestr,pat,figname): 
    xtickrange = {'P10':np.arange(0,22,10),
                  'P09':np.arange(0,1,0.5),
                  'P14':np.arange(0,2,1),
                 'P01':np.arange(0,5,2),
                 'P13':np.arange(0,2,1),
                 'P42':np.arange(0,4,2),
                 'P46':np.arange(0,15,5)}
    xtickpat = xtickrange[pat]
  
    # fetch renamed sample names 
    sample_rename_dict = get_sample_hbar_rename_dict()
    sample_rename = [sample_rename_dict[s] for s in samplestr]
    
    psize = get_barplot_size(pat,'horizontal')
    
    fig = plt.figure(figsize=psize, dpi=300)
    xval=np.flipud(np.arange(len(obs)))
    barlist=plt.barh(xval,
                     obs,
                     0.5,
                     align='center',
                     alpha=0.8,
                    xerr=stds)
    plt.ylim((-0.5,1.5))
    for o in range(len(obs)): 
        barlist[o].set_color(ALL_COLORS_SCALED[o])
    axis_font = {'fontname':'Arial', 'size':'20'} # 3/28/20: ALL THAT MATTERS
    plt.xticks(xtickpat,xtickpat,**axis_font)
    plt.yticks(xval,sample_rename,**axis_font)
    plt.rcParams['xtick.labelsize'] = 10 #14 
    plt.xlabel(r'$<d_{MRCA}> $',**axis_font)
    ax = plt.axes()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
#     plt.show()
    plt.savefig(figname,bbox_inches='tight')
#     plt.close(fig) 
    
patient = 'P10'
treeh, newick = load_patient_tree(patient)
samplestr, _ = get_unique_sample_strings(newick)

obs_means, obs_stds = get_mean_MRCA(treeh,samplestr)

obs_means_keep = [o for o,s in zip(obs_means, samplestr) if 'SP' in s]
obs_stds_keep = [e for e,s in zip(obs_stds, samplestr) if 'SP' in s]
samples_keep = [s for s in samplestr if 'SP' in s]
        
outfig_bar = 'RTTavg_%s_color.pdf' %(patient)
plot_hbar(obs_means_keep,obs_stds_keep,samples_keep,patient,outfig_bar)
```

```python
import pandas as pd

def get_MIC_plot_attr(patient):
    patient_plot_width = {'P13':20, #90, 
                          'P14':40, #170,
                          'P42':50,
                          'P10':250,
                            'P46':50,
                          'P45':150,
                         'P09':40,
                         'P01':60,
                         'P05':140}
    patient_plot_height = {'P45':200, #90, 
                           'P13':400,
                          'P14':400, #170,
                          'P42':400,
                          'P10':400,
                         'P46':400,
                         'P09':400,
                         'P01':400,
                         'P05':250}
    patient_tip_font = {'P13':"7px",
                        'P14':"7px",
                        'P45':"8px", #"8px"
                       'P42':"7px",
                       'P10':"8px",
                       'P46':"7px",
                       "P09":"6px",
                       'P01':"6px",
                       'P05':"6px"} #"8px"
    try: 
        wid = patient_plot_width[patient]
        hgt = patient_plot_height[patient]
        fsize = patient_tip_font[patient]
    except KeyError: 
        wid = 400
        hgt = 500
        fsize = "9px"
    return wid, hgt, fsize

def load_patient_MICs(patname,drugname,outidx):
    filename = patname+'_MIC_heatmap_subset.csv'
    df = pd.read_csv(filename,na_values=np.NAN)
    
    # add dummy outgroup data
    df = df.append(pd.Series(df.shape[1]*[np.NAN]),ignore_index=True)
    drug_concs = df[drugname].tolist()
    
    return drug_concs

def plot_patient_phylogeny_MIC(pat): 
    tfile = '%s/%s_singleSNP.tree' %(pat,pat)

    newick = get_tree(tfile)
    treeh = toytree.tree(newick)
    treeh.ladderize()
    treeh.root("Outgroup")
    treeh.rotate_node(names='Outgroup')

    # recolor
    samplestr, samplecolors = get_unique_sample_strings(newick)
    tipcolors = get_tip_colors(samplecolors, treeh)
    reversedtipcolors = tipcolors
    # reversedtipcolors.reverse()
    nodecolors = get_node_colors(samplecolors,treeh)

    nls = ["" if node not in treeh.get_tip_labels() else node 
           for node in treeh.get_node_dict().values()]

    tree_renamed = rename_tree_leaves(treeh)
    
    ntips = tree_renamed.ntips
    rtre = tree_renamed
    outgroup_idx = rtre.get_tip_labels().index('Outgroup')
    
    # load MICs
    mic_cefep = load_patient_MICs(pat,'Cefepime',outgroup_idx)
    mic_ceftaz = load_patient_MICs(pat,'Ceftazidime',outgroup_idx)
    mic_lev = load_patient_MICs(pat,'Levofloxacin',outgroup_idx)
    mic_mero = load_patient_MICs(pat,'Meropenem',outgroup_idx)

    # start plotting
    plot_width, plot_height, tip_font_size = get_MIC_plot_attr(pat)
    
    # setup a toyplot Canvas and Cartesian grid with two axes
    basepos = 20
    spacer = 5
    mic_width = 30
    canvas = toyplot.Canvas(width=(basepos+plot_width+5*spacer+4*mic_width), height=plot_height)
    ax0 = canvas.cartesian(bounds=(basepos, basepos+plot_width, '10%', '90%'))
    ax1 = canvas.cartesian(bounds=(basepos+plot_width+1*spacer, basepos+plot_width+mic_width+1*spacer, '10%', '90%'))
    ax2 = canvas.cartesian(bounds=(basepos+plot_width+1*mic_width+2*spacer, basepos+plot_width+2*spacer+2*mic_width, '10%', '90%'))
    ax3 = canvas.cartesian(bounds=(basepos+plot_width+2*mic_width+3*spacer, basepos+plot_width+3*spacer+3*mic_width, '10%', '90%'))
    ax4 = canvas.cartesian(bounds=(basepos+plot_width+3*mic_width+4*spacer, basepos+plot_width+4*spacer+4*mic_width, '10%', '90%'))

    figout2 = rtre.draw( axes=ax0,
                    orient='right',
                    width=plot_width, height=plot_height,
                    use_edge_lengths=True, 
                    tip_labels_align=False,
                    tip_labels_colors=reversedtipcolors,
                    tip_labels=True,
                    tip_labels_style={                                                                                                                                                                                      
                        "font-size":tip_font_size,
                        "font-weight":"bold",
                        "-toyplot-anchor-shift": "3px"
                        },
                    node_labels=False, 
                    node_sizes=0,
                    node_labels_style={ "baseline-shift": "15px", 
                                        "-toyplot-anchor-shift": "15px",
                                        "font-size": "10px"}, 
                    edge_style={"stroke-width":1}
                    );
    ax0.show = False

    ax1.bars(
        np.arange(ntips),
        mic_lev,
        along='y',
        color='#A9A9A9',
    );
    ax1.show = False
    ax1.y.show = False
    ax1.x.domain.max = 0
    ax1.x.domain.max = 4
    ax1.x.ticks.show = False

    ax2.bars(
        np.arange(ntips),
        mic_mero,
        along='y',
        color='#A9A9A9',
    );
    ax2.show = False
    ax2.y.show = False
    ax2.x.domain.min = 0
    ax2.x.domain.max = 16
    ax2.x.ticks.show = False

    ax3.bars(
        np.arange(ntips),
        mic_cefep,
        along='y',
        color='#A9A9A9',
    );
    ax3.show = False
    ax3.y.show = False
    ax3.x.domain.min = 0
    ax3.x.domain.max = 64
    ax3.x.ticks.show = False

    ax4.bars(
        np.arange(ntips),
        mic_ceftaz,
        along='y',
        color='#A9A9A9',
    );
    ax4.show = False
    ax4.y.show = False
    ax4.x.domain.min = -0.25
    ax4.x.domain.max = 64
    ax4.x.ticks.show = False
```
