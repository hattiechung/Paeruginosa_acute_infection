```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# filename = '/Users/hchung/Dropbox/projects/kishony-acute-infection/phenotypes/SI_phenotype_data_CM_10.01.19.xlsx'
filename = 'phenotypic_assays/Extended_Data_phenotypic_assays.xlsx'
```

# Twitching

```python
phenotype = 'Twitching' #'SerumAssay' 
df = pd.read_excel(filename, sheet_name=phenotype)
```

```python
sns.set(rc={'figure.figsize':(6,4)},font_scale=1.5)
sns.set_style('white')
sns.despine()
sns.swarmplot(x='Isolate',y='Twitching diameter (cm)',data=df,color='w',edgecolor='gray',linewidth=2,size=8)
fh = sns.barplot(x='Isolate',y='Twitching diameter (cm)',data=df,ci='sd',estimator=np.median,
            capsize=0.3,facecolor=(1, 1, 1, 0),edgecolor="k")
plt.xticks(rotation=90)
fh.get_figure().savefig('%s.pdf' %phenotype,bbox_inches='tight')
```

# WbpL Serum Assay

```python
phenotype = 'SerumAssay' #'SerumAssay' 
df = pd.read_excel(filename, sheet_name=phenotype)
```

```python
sns.set(rc={'figure.figsize':(4,4)},font_scale=1.5)
sns.set_style('white')

sns.swarmplot(x='Isolate',y='CFU/mL',hue='Treatment',data=df,dodge=True,
              order=['PAO1','C-20','C-8','C-23 (wbpL)','C-10 (wbpL)','C-2 (wbpL)'],
              hue_order=['50% serum'],
                size=3, color='w',edgecolor='gray',linewidth=2)
fh = sns.barplot(x='Isolate',y='CFU/mL',hue='Treatment',data=df,
                 order=['PAO1','C-20','C-8','C-23 (wbpL)','C-10 (wbpL)','C-2 (wbpL)'],
              hue_order=['50% serum'],estimator=np.median,
            capsize=0.3,facecolor=(1, 1, 1, 0),edgecolor="k")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
fh.set_yscale('log')
sns.despine()
fh.set_ylim((1,100000000))

fh.get_figure().savefig('%s.pdf' %phenotype,bbox_inches='tight')
```
